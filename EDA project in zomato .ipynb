## EDA-6 : IDENTIFY the outlier
### Q1 ( first quartile, 25th)
### Q3 (third Quartile, 75th)
### IQR (Interquartile range) = Q3 - Q1
### Lower bound = Q1 -1.5 * IQR
### UPPER bound = Q3 + 1.5 * IQR

```python
df10 = df.copy()
sns.boxplot(x = 'votes', data = df10)
```

```python
Q1 = df.votes.quantile(.25)
Q3 = df.votes.quantile(.70)

print(Q1)
print(Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR
print(Lower_bound)
print(Upper_bound)
```

```python
outliers = df['votes'][(df.votes < Lower_bound) | (df.votes > Upper_bound)]
print(outliers)
```

```python
outliers.count()
```

# Data Cleaning-4 : Trim the outliers

```python
df11 = df.copy()
df11 = df11[(df.votes >= Lower_bound) & (df.votes <= Upper_bound)]
```

```python
df11.votes.unique()
```
